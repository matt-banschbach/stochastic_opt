```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 15 # decision variables
m = 10 # number of constraints per node
p = 7 # number of linking variables
```

```python
def nonanticipativity(H, k):
    return n * (  (k**(H-1)) -1)
```

```python
def num_columns(H, k):
    if k == 1:
        return n * H
    else:
        return n * ((k**H -1) / k-1)
```

```python
def num_rows(H, k):
    if k == 1:
        return m*H + p*(H-1)
    else:
        return (m * ((k**H -1) / k-1)) + (p * ((k**H - k) / k-1))
```

```python
# Fix H
H_val = 5

k_values = np.arange(1, 8) # Plot for k from 1 to 7

# Calculate the number of constraints/variables for each k
nonant_values = [nonanticipativity(H_val, k) for k in k_values]
col_values = [num_columns(H_val, k) for k in k_values]
row_values = [num_rows(H_val, k) for k in k_values]

plt.figure(figsize=(10, 6))

plt.plot(k_values, nonant_values, label=f'Nonanticipativity (n={n})', marker='o')
plt.plot(k_values, col_values, label=f'Columns (n={n})', marker='o')
plt.plot(k_values, row_values, label=f'Rows (m={m}, p={p})', marker='o')

plt.yscale('log')
plt.xlabel('Number of Children per Node (k)')
plt.ylabel('Count (log scale)')
plt.title(f'Growth of Constraints and Variables with k for H={H_val}')
plt.legend()
plt.grid(True, which="both", linestyle='--')
plt.xticks(k_values)
plt.show()
```

```python
k_val = 4

H_values = np.arange(1, 11) # Plot for k from 1 to 10

nonant_values = [nonanticipativity(H, k_val) for H in H_values]
col_values = [num_columns(H, k_val) for H in H_values]
row_values = [num_rows(H, k_val) for H in H_values]

plt.figure(figsize=(10, 6))

plt.plot(H_values, nonant_values, label=f'Nonanticipativity (n={n})', marker='o')
plt.plot(H_values, col_values, label=f'Columns (n={n})', marker='o')
plt.plot(H_values, row_values, label=f'Rows (m={m}, p={p})', marker='o')

plt.yscale('log')
plt.xlabel('Number of Stages (H)')
plt.ylabel('Count (log scale)')
plt.title(f'Growth of Constraints and Variables with H for k={k_val}')
plt.legend()
plt.grid(True, which="both", linestyle='--')
plt.xticks(k_values)

plt.show()
```
